```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings, gc
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('data_mining_project.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#Dropping Columns having more than 50% null values.
i=0
for col in df.columns:
    if (df[col].isnull().sum()/len(df[col])*100) >=50:
        print("Dropping column", col)
        df.drop(labels=col,axis=1,inplace=True)
        i=i+1
        
print("Total number of columns dropped in train dataframe", i)
```

```python
#Replacing Categorical Variable null values with its mode.
for i in ['D_126','D_120','D_117','D_116','D_114','B_38','B_30','D_63','D_64','D_68']:
    df[i][df[i].isnull()] = df[i].dropna().mode()[0]
    
```

```python
def lower_limit(i):
    x = df[i].quantile(0.25) - 1.5*(df[i].quantile(0.75)-df[i].quantile(0.25))
    indexes = df[df[i]<x].index.to_list()
    df[i][indexes] = x
```

```python
def upper_limit(i):
    x = df[i].quantile(0.75) + 1.5*(df[i].quantile(0.75)-df[i].quantile(0.25))
    indexes = df[df[i]>x].index.to_list()
    df[i][indexes] = x
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
X7= df_pca.iloc[:,np.r_[0:44]]
y7= df_pca.iloc[:,44]
```

```python
from sklearn.model_selection import train_test_split
X7_train, X7_test, y7_train, y7_test = train_test_split(X7, y7, test_size=0.2, random_state=42, stratify=y7)
```

```python
x7_train = ohe.fit_transform(X7_train[['D63','D64']])
x7_test = ohe.transform(X7_test[['D63','D64']])
```

```python
X7_train_d=X7_train.drop(['D63','D64'],axis=1)
X7_test_d=X7_test.drop(['D63','D64'],axis=1)
```

```python
X7_train_trans=np.hstack((X7_train_d.values,(x7_train)))
X7_test_trans=np.hstack((X7_test_d.values,(x7_test)))
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X7_train=ss.fit_transform(X7_train_trans)
X7_test=ss.fit_transform(X7_test_trans)
```

```python
sm=SMOTE(random_state=12,sampling_strategy='minority')
X7_train_sm, y7_train_sm = sm.fit_resample(X7_train,y7_train)
```

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier(
    n_estimators=300,          # Number of boosting rounds
    learning_rate=0.1,         # Step size during each boosting iteration
    max_depth=3,               # Maximum depth of a tree
    min_child_weight=1,        # Minimum sum of instance weight needed in a child
    subsample=1.0,             # Fraction of samples used for fitting the trees
    colsample_bytree=1.0,      # Fraction of features used for fitting the trees
    gamma=.2,                   # Minimum loss reduction required to make a further partition
    reg_alpha=1,               # L1 regularization term
    reg_lambda=1,              # L2 regularization term
    scale_pos_weight=1,        # For imbalanced datasets
    objective='binary:logistic',  # Learning task and objective function
    eval_metric='logloss'      # Evaluation metric
)
```

```python
xgb_model.fit(X7_train_sm,y7_train_sm)
```

```python
y_pred=xgb_model.predict(X7_test)
y_pred
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y7_test,y_pred))
print(metrics.classification_report(y7_test,y_pred))
```

```python
y_pred_t=xgb_model.predict(X7_train_sm)
print(metrics.classification_report(y7_train_sm,y_pred_t))
```
